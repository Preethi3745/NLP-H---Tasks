```python
import nltk

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag, FreqDist


stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

text = input("Enter text: ")

# tokenization
tokens = word_tokenize(text)

# Stopword Removal
filtered = []
for word in tokens:
    if word.lower() not in stop_words and word.isalpha():
        filtered.append(word)

# stemming
stemmed = []
for word in filtered:
    stemmed.append(stemmer.stem(word.lower()))

# lemmatization
lemmatized = []
for word in filtered:
    lemmatized.append(lemmatizer.lemmatize(word.lower()))

# tagging
tagged = pos_tag(lemmatized)

# frequency
freq_dist = FreqDist(lemmatized)

# Output
print("Tokenization:", tokens)
print("Without Stopwords:", filtered)
print("After Stemming:", stemmed)
print("After Lemmatization:", lemmatized)
print("Tagged:", tagged)

print("\nFrequency Distribution:")
for word, freq in freq_dist.items():
    print(word, ":", freq)
```
