```python
from collections import Counter
import nltk
import string
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

text=input("enter news articles: ")
text=text.lower()

for t in string.punctuation:
    text=text.replace(t,"")
words=word_tokenize(text)

stop_words=set(stopwords.words('english'))
filtered_words=[w for w in words if w not in stop_words]

total_words=len(filtered_words)

unique_words=set(filtered_words)

freq=Counter(filtered_words)
print("Word Frequncies")
print(freq)


for word,count in freq.most_common(5):
    print(word,":",count)
ttr=len(unique_words)/total_words
print("\nType Token Ratio=",round(ttr,2))

dominant_topic=freq.most_common(1)
if dominant_topic:
    print("\nMost Dominant Topic: ",dominant_topic[0][0])

word_list=list(freq.keys())
counts_list=list(freq.values())

plt.figure(figsize=(8,5))
plt.bar(word_list, counts_list)
plt.title("NLP Word Frequency Analysis")
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.xticks(rotation=90)
plt.show()

```
