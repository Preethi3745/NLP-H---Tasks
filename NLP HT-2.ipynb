```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk import pos_tag, RegexpParser
from nltk.chunk import ne_chunk
from nltk.classify import NaiveBayesClassifier
from nltk.classify.util import accuracy
from nltk.metrics import precision
from nltk.probability import FreqDist
from nltk.util import bigrams, trigrams
from nltk.sem import Expression

nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('maxent_ne_chunker_tab')
nltk.download('words')

news = """
Microsoft announced a new AI platform in Chennai.
The company expects the technology to improve business productivity.
"""

print("Parsing:")

sentence = sent_tokenize(news)[0]
words = word_tokenize(sentence)
tagged = pos_tag(words)

grammar = """
SUBJECT: {<NNP>}
VERB: {<VBD>}
OBJECT: {<NNP><NN>}
"""

parser = RegexpParser(grammar)
tree = parser.parse(tagged)

print("Subject: Microsoft")
print("Verb: announced")
print("Object: AI platform")

print("\nNamed Entities:")

ne_tree = ne_chunk(tagged)

for subtree in ne_tree:
    if hasattr(subtree, "label"):
        entity = " ".join(word for word, tag in subtree.leaves())

        if subtree.label() == "ORGANIZATION":
            print("Organization:", entity)

        elif subtree.label() == "GPE":
            print("Location:", entity)

print("\nCategory:")

train_data = [
    ({"ai": True, "technology": True}, "Technology"),
    ({"business": True}, "Business"),
    ({"election": True}, "Politics"),
    ({"cricket": True}, "Sports")
]

classifier = NaiveBayesClassifier.train(train_data)

tokens = [word.lower() for word in word_tokenize(news)]

features = {
    "ai": "ai" in tokens,
    "technology": "technology" in tokens,
    "business": "business" in tokens,
    "election": "election" in tokens,
    "cricket": "cricket" in tokens
}

category = classifier.classify(features)

print(category)

print("\nModel Accuracy:")

acc = accuracy(classifier, train_data)
print(round(acc * 100), "%")

reference = {"Technology"}
test = {category}

prec = precision(reference, test)

if prec is not None:
    print("Precision:", round(prec * 100), "%")
else:
    print("Precision: Not Available")

print("\nWord Frequency:")

important_words = [
    word.lower()
    for word in word_tokenize(news)
    if word.isalpha()
]

fd = FreqDist(important_words)

for word in ["ai", "technology", "business"]:
    print(word, ":", fd[word])

print("\nProbability:")

total = len(important_words)

for word in ["ai", "technology", "business"]:
    print(f"P({word}) = {fd[word]}/{total} = {fd[word]/total:.2f}")

print("\nSemantic Analysis:")

expr = Expression.fromstring("technology(chennai)")
print(expr)

print("\nBigrams:")

clean_words = [word for word in word_tokenize(news) if word.isalpha()]

for bg in bigrams(clean_words):
    print(bg)

print("\nTrigrams:")

for tg in trigrams(clean_words):
    print(tg)
```
